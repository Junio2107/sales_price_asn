```python
df = pd.read_excel(r'C:\Users\Junio\Desktop\teste_agrupamento.xlsx')
```

```python
df_group = pd.DataFrame(df.groupby('X')['Y'].mean()).reset_index()
print(df_group)
```

```python
df_dummy = pd.get_dummies(df['X']).astype(int)
df_dummy
```

```python
df_dummy = pd.concat([df,df_dummy], axis=1)
df_dummy
```

```python
features = df_dummy.columns.drop(['Y'])
formula = 'Y ~ + A + B + C'
reg_mult = smf.ols(formula=formula,data=df_dummy).fit()
print(reg_mult.summary())
```
